```python
import pandas as pd


import os

def formatear_numero(numero):
    return f"{numero:,.0f}".replace(',', 'X').replace('.', ',').replace('X', '.')

def formatear_porcentaje(porcentaje):
    return f"{porcentaje:.2f}".replace('.', ',')

def generartexto(cod_producto, df, dfprov, adestino):
    # Filtrar el DataFrame para obtener solo las filas correspondientes al código de producto
    df_producto = df[df['k_4dig'] == cod_producto]
    
    # Verificar si el DataFrame filtrado está vacío
    if df_producto.empty:
        return "Producto no encontrado."
    
    # Obtener la descripción del producto
    descripcion = df_producto['desc_x'].iloc[0]
    
    # Obtener los principales exportadores y sus valores
    exportadores = df_producto[['country_name', 'v']]
    
    # Construir el texto con la información disponible
    texto = f"En 2022, los principales exportadores de ({cod_producto}) {descripcion} fueron:\n"
    
    exportadores_texto = ', '.join([f"{row['country_name']} (US${formatear_numero(int(row['v']))})"
                                    for index, row in exportadores.iterrows()])
    
    porcentaje_total = df_producto['porcentaje_mundial_expo_producto_y'].iloc[0]
    texto += f"{exportadores_texto}, representando en conjunto el {formatear_porcentaje(porcentaje_total)}% de las exportaciones mundiales.\n\n"
    
    # Filtrar el DataFrame de provincias
    prod_provi = dfprov[dfprov['k_4dig'] == cod_producto]
    
    # Convertir la columna 'FOB_PROVINCIA_PRODUCTO_TOTAL' a int64
    prod_provi['FOB_PROVINCIA_PRODUCTO_TOTAL'] = prod_provi['FOB_PROVINCIA_PRODUCTO_TOTAL'].astype('int64')
    
    # Obtener la suma de las exportaciones por provincia
    sumaprovincias = prod_provi['FOB_PROVINCIA_PRODUCTO_TOTAL'].iloc[0]
    
    # Construir el texto para las provincias
    provincias = (f"Dentro de Argentina, las provincias exportadoras y sus respectivas contribuciones "
                  f"porcentuales al total de exportaciones, fueron:\n")
    
    texto += provincias
    
    provinciaspuntuaes = ', '.join([f"{row['DESCPROVI']} con el {formatear_porcentaje(row['porcentaje_del_total'])}%"
                                    for index, row in prod_provi.iterrows()])
    
    texto += provinciaspuntuaes + ".\n\n"
    
    # Filtrar y ordenar el DataFrame de destinos
    adestino = adestino.sort_values(by=['k_4dig', 'v'], ascending=[True, False])
    adestino = adestino.groupby('k_4dig').head(3).reset_index(drop=True)
    dfdestino = adestino[adestino['k_4dig'] == cod_producto]
    
    # Construir el texto para los destinos
    paisesdestino = ', '.join([f"{row['country_name']} ({formatear_numero(int(row['valordestinos']))} US$)"
                               for index, row in dfdestino.iterrows()])
    
    datoporcentaje = dfdestino['v'].iloc[0]/(df[(df['k_4dig'] == cod_producto) & (df['i'] == 32)]['v'].sum())*100
    
    datoporcentaje = datoporcentaje.astype('int64')
    
    texto += f"Los principales destinos de exportación de Argentina de este producto fueron:\n{paisesdestino}, los cuales representan el {datoporcentaje}% de las exportaciones de Argentina de {descripcion}."
    
    return texto




valores_unicos_k_4dig = tabla_final_4digs['k_4dig'].unique()



def automatizar_textos(lista_valores_unicos):
    output_path = os.path.join(os.getcwd(), 'output_textos.txt')
    with open(output_path, 'w', encoding='utf-8') as file:
        for valor in lista_valores_unicos:
            texto_generado = generartexto(valor, tabla_final_4digs, top_3_expo_provinciales_producto, dataframeconnombresdedestinoquenoentiendo)
            file.write(texto_generado + '\n\n')
                


automatizar_textos(valores_unicos_k_4dig)

```
