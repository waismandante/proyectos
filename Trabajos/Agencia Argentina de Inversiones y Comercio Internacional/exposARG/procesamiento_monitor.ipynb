```python

def cantidad_de_digitos(numero):
    # Calcular la cantidad de dígitos de un número
    return len(str(numero))

def cantidad_de_miles_de_millones(numero):
    digitos = cantidad_de_digitos(numero)
    
    if digitos == 10:
        # Si tiene 10 dígitos, tomar el primer dígito
        return int(str(numero)[0])
    elif digitos == 11:
        # Si tiene 11 dígitos, tomar los dos primeros dígitos
        return int(str(numero)[:2])
    else:
        return "El número no tiene 10 o 11 dígitos."

# Ejemplo de uso
resultado = cantidad_de_miles_de_millones(exportacion_mensueal)
print(f"Cantidad de miles de millones: {resultado}")
```
